**17- Splitting string into multiplle coloumns**

```python
df=pd.DataFrame({'name':['Muhammad Sohail','Jibran marwat','Muhammad Umar'],
                 'Location':['Mansehra, Pakistan',' marwat, Pakistan','Oghi, Germany'],
                 'Degree':['SE','SE','AI']})
df
```

```python
df[['First_Name','Last_Name']]=df.name.str.split(' ',expand=True)
df
```

```python
df[['City','Country']]=df.Location.str.split(', ',expand=True)

df
```

```python
df=df[['First_Name','Last_Name','City','Country']]
df
```
